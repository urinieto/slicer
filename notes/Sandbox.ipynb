```python
import colorednoise as cn
import matplotlib.pyplot as plt
import librosa
import numpy as np

%matplotlib inline
```

```python
X = librosa.feature.melspectrogram(y=cn.powerlaw_psd_gaussian(1, size=(22050)),
                                   sr=22050,
                                   n_mels=80,
                                   n_fft=1024,
                                   hop_length=512,
                                   fmin=0,
                                   fmax=None)
```

```python
plt.imshow(librosa.power_to_db(X, ref=np.max))
```
